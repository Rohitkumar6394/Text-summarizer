```python
from transformers import pipeline

summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

print("Model Loaded")
```

```python
MEMORY_FILE = "summary_memory.txt"

open(MEMORY_FILE, "a").close()

print("Memory File Created")
```

```python
def read_memory():

    with open(MEMORY_FILE, "r", encoding="utf-8") as f:
        return f.read()


def write_memory(text):

    with open(MEMORY_FILE, "w", encoding="utf-8") as f:
        f.write(text)
```

```python
def add_conversation(user_text):

    old_memory = read_memory()

    new_memory = old_memory + "\n" + user_text

    return new_memory
```

```python
def summarize_memory(text):

    if len(text.split()) < 30:
        return text

    result = summarizer(
        text,
        max_length=100,
        min_length=20,
        do_sample=False
    )

    return result[0]["summary_text"]
```

```python
def memory_agent(user_input):

    current_memory = add_conversation(
        f"User: {user_input}"
    )

    summary = summarize_memory(current_memory)

    write_memory(summary)

    return summary
```

```python
while True:

    text = input("You: ")

    if text.lower() == "exit":
        break

    summary = memory_agent(text)

    print("\nCurrent Memory Summary:")
    print(summary)
    print("-" * 50)
```
