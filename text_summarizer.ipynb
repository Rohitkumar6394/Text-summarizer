```python
from langchain_core.prompts import PromptTemplate
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline
```

```python
llm = HuggingFacePipeline(
    model_id="google/flan-t5-base",
    pipeline_kwargs={
        "task": "text2text-generation",
        "max_new_tokens": 256
    }
)
```

```python
prompt = PromptTemplate(
    input_variables=["text"],
    template="summarize: {text}"
)
```

```python
chain = prompt | llm
```

```python
text = """
Artificial Intelligence (AI) is a branch of computer science that aims to create intelligent machines
that can think, learn, and perform tasks like humans. It is widely used in healthcare, education,
transportation, and many other industries. Machine learning is a subset of AI that allows systems
to learn from data and improve over time without being explicitly programmed.
"""
```

```python
summary = chain.invoke({'text': text})

print("рџ“Њ Original Text:\n", text)
print("\nрџ§  Summary:\n", summary)
```
